```python
#Advanced 1 
#Asking how many inputs(assets) there are
numAssets = int(input('Enter the number of assets:'))
#Setting the intial value to 0 here. This give us a place to start from when doing the sum of all assets.
tValue = 0.0
#Here we are setting the range to equal i+1 until it loop gets to numAsset input. This loop lets values and count be assigned to the assets.
for i in range(1, numAssets + 1):
    assetP = float(input("Enter Asset {} value:".format(i)))
    tValue += assetP
#This condition allows for two different outputs depending on if you have a single asset, or more than one asset. 
if numAssets == 1 :
    print("The value of the {} assets is ${:.2f}".format(numAssets,tValue))
else:
    print("The total value of the {} assets is ${:.2f}".format(numAssets,tValue))

```

```python
#Advanced #2
#First thing first we need to enter the conversion factors from miles into inches and then imperial to metric systems, used ChatGPT for quick conversion metrics. Same idea as law of one price, need like denomenator in order to do simple algebra. First 3 conversions assume 1X of something to many
Mile_Inch = 63360
Yard_Inch = 36
Foot_Inch = 12
#Used to using 2.54 as the conversion factor from chemistry, same as 1/0.3937
#Second 3 here are assuming many of something to 1X of something.
Inch_CM = 2.54
CM_Meter = 100
CM_KM = 100000
#Get user inputs here for each of imperical values
m = float(input('Enter miles:'))
y = float(input('\nEnter yards:'))
f = float(input('\nEnter feet:'))
inch = float(input('\nEnter inches:'))
#Next convert everything into the same denomantor, in this case it will be inches and then sum it all together after conversions.
totalInch = (m*Mile_Inch)+(y*Yard_Inch)+(f*Foot_Inch)+(inch)
#Next to get to from imperial to metric system we convert inches to centemeters with the given coversion rate 1 inch = 2.54cm (0.3937 inch/cm)
totalCM = (totalInch * Inch_CM)
#Now that we have total CM we can use integer divison to break down the total into km, meters, and cm with remainders. We also use remainder division to take whats left and assign it to the next tier, in this case meters and then cm
km = int(totalCM // CM_KM)
remainder = totalCM % CM_KM
meters =int(remainder // CM_Meter)
remainder = remainder % CM_Meter
cm = round((remainder),1)
#Print Final Answer
print("The metric length is:\n{} kilometers\n{} meters\n{} centimeters".format(km,meters,cm))


#I was stuck on this one for awhile because my output did not equal what yours had in the lab 1 assignment. I noticed the footnote and with 1/.3937 = 2.5400508 it only makes sense that your answer was a touch higher than mine.

```

```python
#Advanced 3 
#First we need to define the fee rate of 6.9%. Here we type the input equivlents percent into a number
transaction_fee =0.069
#Second were getting the inputs regarding the type of transaction the user hopes to exchange for. In this case they need to input the driver of the script which is the coversion rate and then which currency they would like to trade from, in this case dollars to euros or euros to dollars.
exchange_rate = float(input("Please enter the exchange rate (1 USD = ? EUR):"))
#Added a .lower() in here so incase someone typed "Dollars" or "Euros" is still would execute correctly. I thought about phones especailly and the automatic capitalization of the first key in a text.
currency = input("Do you wish to exchange dollars or euros?").lower() 
#First we need to make the condition logic which way we are converting. The if statement allows for us to seperate the different functions by the type of transaction occuring. This first example here is dollars to euros. Firstly the user inputs the amount of US dollars they want to exchange.
if currency == "dollars" :
    dollars = float(input("Please enter the number of US Dollars: "))
#Once we have the input of US dollars we can start the algebra to exchange it to euros, in this case every US dollar is worth .92 Euro. Once its converted we then can take multiply the total amount by the 6.9% transaction fee. 
    euros = dollars*exchange_rate
    fee = euros * transaction_fee
#After we find the fee, we can subtract that amount from our total amount of euros to see our new total after the transaction had occured.
    total = euros - fee
#Finally for dollars we have the outputs, each is formatted to have two decimal places and a float value beucase the exchange rate is not perfeclty 1:1 we can expect to see fractions in our outputs.  We also use .format() function to call the related totals down into the print outputs. 
    print("\nThe total exchange value is {:.2f} Euros".format(euros))
    print("The transactionf fee is {:.2f} Euros".format(fee))
    print("\nYou eill recieve a total of {:.2f} Euros".format(total))
#Here we run through basically the exact same function as above, but with one catch. If were going from euros to dollars we have to divide by the excahnge rate rather then multiply. 
elif currency== "euros" :
    euros = float(input("Please enter the number of Euros: "))
    dollars = euros / exchange_rate
    fee = dollars * transaction_fee
    total = dollars - fee
#Here we have the expected printed outputs for our euro to dollar exchange rate. Again it is formatted with 2 decimal places and a float value because the exchange rate is not 1:1. We also use .format() function to call the related totals down into the print outputs. 
    print("\nThe total exchange value is ${:.2f}".format(dollars))
    print("The transaction fee is ${:.2f}".format(fee))
    print("\nYou will recieve a toal of ${:.2f}".format(total))
else:
#Here we have the else statement, which is where the script goes if not 'dollars' or 'euros' is inputted in the sixth line of code or second asked question. 
    print("That is not a valid selection.")
    
```

```python
#Advanced #4
#The first thing were going to do is get the user inputs. First balance, second # of accounts, third is minnesota resident status. 
balance = float(input("Enter account balance: "))
accounts = int(input("How many accounts do you have with us?" ))
status = str(input('Are you a Minnesota resident: "yes" or "no" ? ' )).lower()
#Next we are setting up some conditions to represent a boolean of true or false for each of the three user inputs. (T or F)
condition1 = status == "yes"
condition2 = balance > 10000
condition3 = accounts >= 3
#Next we are going to do a quick sum of the truths from the conditions above, this is what allows us to satisfy the criteria of 2 conditions must be met for the program to execute with "Approve"
numconditionstrue = sum([condition1,condition2,condition3])

#Now were going to get into the  if statement. Since we were able to define if the number of criteria was met above more simply with the sum, we dont need to use a crazy nested if statement to code all the possible differnet combinations. Instead the simple if function basically says if 2 or more, print approve. If less, print reject. 
if numconditionstrue >= 2:
    print("\nApprove")
else:
    print("\nReject")
```
