```python
#Basic 1-5
def main():
    x = input("Enter your string: ")
    print("Your string:")

    if x.isalnum():
        print("Is alphanumeric")
    if x.isdigit():
        print("Contains only numbers")
    if x.isalpha():
        print("Contains only letters")
    if x.isupper():
        print("Is in all CAPS")
    if x.islower():
        print("Is in all lowercase")

main()
```

```python
#Intermediate Problem 1
#Firstly we need input numpy as np to write our code as neatly as possible.
import numpy as np
#Next we start our main function - so we can call it later. 
def main():
#Next we need to ask the user the amount of stocks in their portfolio. We also set prices and shares accumulator variables to empty here so they have a starting point. 
    num = int(input('Enter number of stocks in portfolio: '))
    prices = []
    shares = []
#This is a count controlled loop - it is using the input from the user in the "num" input to count how many time it shall run through the loop. For example, the input is two, so it will ask what the price is of the stock twice before moving to our next count controlled loop.    
    for i in range(1,num+1):
        #Input for price of stock from user
        p = float(input('Enter price of stock {}:\n'.format(i)))
                #This function here appends the prices list that we set to 0 earlier with the accumulator variables, the newly inputted p value from the user is added to the list prices.
        prices.append(p)
#This is a count controlled loop - it is using the input from the user in the "num" input to count how many time it shall run through the loop. This allows us to have seperate values for each stock. 
    for i in range(1,num+1):
        s = int(input('Enter the # shares of stock {}:\n'.format(i)))
        #This function here appends the shares list that we set to 0 earlier with the accumulator variables, the newly inputted s value from the user is added to the list shares.
        shares.append(s)
     #Next we call numpy to make arrays of our new inputs. In essence this array is a 2x2 box because we have 2 inputs for each list.   
    price = np.array(prices)
    shares = np.array(shares)
    #We can then call the numpy function "multiply" to multiply the price array by the shares array. That is equal to our variable "value"
    value=np.multiply(price,shares)
    #Now that we have values for 1 and 2, we can use the sum function to sum together each of the values calculated above from "Multiply"
    total = np.sum(value)
    #This count controlled loop lets us have outputs for each index value of i. Multiple outputs based on orgianl input for how many stocks we have in the portfolio. 
    for i in range(num):
        print("Your holdings for stock {} = ${}".format(i + 1, int(value[i])))
    
    print("The total value of your portfolio is: ${:.2f}".format(total))
#Call the main function!
main()
```

```python
#Intermediate Problem 2 
#Import numpy first as variable np
import numpy as np
#Start by defining out main function 
def main():
    #Input from user about how many stocks are in the portfolio
    num = int(input('Enter number of stocks in portfolio: '))
    #Assign accumulator variables to empty
    prices = []
    shares = []
    names=[]
    #First count controlled loop that appends(adds) the array "names" for the input that the user provides for stocks in the portfolio based on their num input.
    #The +1 in the formatting assures that we get the correct amount of inputs. 
    for i in range(num):
        n = input('\nEnter the name of stock {}:\n\n'.format(i+1))
        names.append(n)
    #second count controlled loop that appends(adds) the array "prices" for the input that the user provides for stocks in the portfolio based on their num input.
    #The formatting here is calling on the names index that was created by appending the "names" array above. Allows the user to see which stock the program is asking for rather then have a number value.
    for i in range(num):
        p = float(input('\nEnter price of stock {}:\n\n'.format(names[i])))
        prices.append(p)
    #third count controlled loop that appends(adds) the array "shares" for the input that the user provides for stocks in the portfolio based on their num input.
    #The formatting here is calling on the names index that was created by appending the "names" array above. Allows the user to see which stock the program is asking for rather then have a number value.
    #This loop also allows for the multiplication to be ran for a multiple number of stocks (num input)
    for i in range(num):
        s = int(input('\nEnter the # shares of stock {}:\n\n'.format(names[i])))
        shares.append(s)
    #next we create our arrays with numpy for each of the user inputs.
    name = np.array(names)
    price = np.array(prices)
    shares = np.array(shares)
    #here we multiply the ararys and assign them each a value. 
    value=np.multiply(price,shares)
    #Sum together both of the values to get our total portfolio worth.
    total = np.sum(value)

    #this loop if for the output. We get individual holdings for each stock provided by the user - formatting the names and values based on the index values in num (0,1).
    #We also get the output of total portfolio wealth. 
    for index in range(num):
        print("\nYour holdings for stock {} = ${:,.0f}".format(names[index],(value[index])))
    
    print("\nThe total value of your portfolio is: ${:,.0f}".format(total))
#Call the main function!
main()
```

```python
#Advanced Problem 1
#Import numpy first as variable np and set accumulator variables to empty
import numpy as np
names = []
values = []
#Start by defining out main function 
def main():
    #Input from user about how many stocks are in the portfolio
    num = int(input('Enter number of stocks in portfolio: '))
    #This count controlled loop is the bulk of the code here. It repeats based on user input.
    for i in range(num):
        #First it prints a expected output formatted for the same amount as the user input.
        print("\n******For asset, {}******".format(i+1))
        #Next is user input, n signifies the assets name. This will also repeat based on user input.
        n = input("\nEnter the name of the asset: ")
        #Here we call the while (condition controlled) loop paired with .isaplpha to check the input n's first character and check for a letter.
        #It continues to ask until that condition is met, once it is met it appends the list "names"
        while not n[0].isalpha():
            print("Asset name must start with a letter!")
            n = input("\nEnter the name of the asset: ")
        names.append(n)
        #Same Concept as above, checking v input with while (condition controlled) loop. This time paired with .isdigit to make sure the input is a number.
        #Once condition is met it formats the input as an intager and adds it to the list "values"
        v = input("\nEnter the value of the asset: ")
        while not v.isdigit():
            print("Asset value must be a number!")
            v = input("\nEnter the value of the asset: ")
        values.append(int(v))
    #Here we are opening the blank file Lab3 to add some data. We are in write mode because 1) file is blank & 2) we need to edit.
    #we do a count controlled loop to account for user input of number of stocks 
    #Formatted to use the index of num to call both of the names and values fot the inputted number of stocks.
    Lab3_File=open("Lab3.txt",'w')
    for i in range(num):
        Lab3_File.write(names[i]+"," + str(values[i])+"\n")
    #We then close the file
    Lab3_File.close()
#And call the main function   
main()
```

```python
#Advanced Problem 1 Part b
#I think numpy is already imported from prior code but being safe
import numpy as np
#Define our main function and assign accumulator variables to empty so we can fill them later.
def main():
    names=[]
    values = []
    #After definitons we can open the file Lab3.txt in read mode so we dont write over the data in the file
    Lab3_File=open("Lab3.txt",'r')
    #Next we use a count controlled loop that repeats for each line of data in the Lab3.txt file
    for line in Lab3_File:
        #First we have to strip the data of any formatting on either end of it, then we split based on the "," inputted from the formatting from earlier code
        parts = line.strip().split(",")
        #Then we append the names list with the first value in new list "parts" and values with the second value in the list "values" (sorry)
        names.append(parts[0])
        values.append(int(parts[1]))
    #Then we close the file
    Lab3_File.close()   
    #After we close the Lab3.txt file we can use our read data and numpy to do some simple calculations.
    #First we make an array of the read values
    value_array = np.array(values)
    #Use numpy functions to get total and average of the values of the assets
    total = np.sum(value_array)
    average = np.average(value_array)
    #Then we print out new outputs
    #Use len to get the number of assets in the names list (had to ask chatgpt how to do that)
    #it then prints all the assets in names
    print("The portfolio has the following {} assets:".format(len(names)))
    for name in names:
        print('\n'+name)
    #Finally we print our totals we got from using numpy to find average and sum
    print("\nThe portfolio is worth ${:,.2f}".format(total))
    print("The portfolio is worth ${:,.2f}".format(average))
#Call the main function    
main()
```
