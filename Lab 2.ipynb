```python
#Basic #1
def Tax(purchase,taxrate):
    return (purchase * taxrate)
    
def main():
    taxrate = 0.08  
    purchase = float(input('Enter the purchase amount: '))
    
    tax=Tax(purchase,taxrate)
    total=tax+purchase
    print ('Purchase amount: $', format(purchase,'.2f'),sep='')
    print ('Sales tax: $', format(tax,'.2f'),sep='')
    print('Total purchase: $',format(total,'.2f'),sep='')

main()
```

```python
#Basic #2
def FV(p,r,t):
    total=p*(1+r/100)**t
    return total

def main(): 
    principal = float(input("Enter principal: "))
    rate = float(input("Enter interest Rate: "))
    time = int(input("Enter # of periods: "))
    futurevalue = FV(principal,rate,time)
    print("The future value is: ${:,.2f}".format(futurevalue))

main()
              
    
```

```python
#Basic #3
import random as rnd

def main():
    flips = int(input("Enter number of coin flips: "))
    heads = 0 #setting Accumulator to 0 so we have a starting poing 

    for _ in range(flips):
        if rnd.random() < 0.5: #This sets heads probability to 50%
            heads += 1
    
    print("You had {} heads out of {} coin flips.".format(heads, flips))

main()
```

```python
#Advanced #2

def main():
    #First thing we need to do is write out the input formulas associated with a annuity calculation. In this case specifically were asked for the intial rate, the interset rate, and monthly payout amount. Here we use float because each of these inputs could possibly have decimal places. \n are used to match the format of the output. 
    initialI = float(input("Please enter your intial investment: "))
    rate = float(input("\nPlease enter annual interest rate (ex. 2.5): "))
    payout = float(input("\nPlease enter the monthly annuity payment: "))

    #The next step we need to do is get the annual rate into a monthly rate, we know this because the third prompt asks the user for the monthly payouts. So we divide the rate by 12 months and by 100 to get a %
    mrate = rate/12/100
    #Next we use aggregator variables to set months and our initial balance to 0 and whatever the user inputs to give the script a starting location. 
    balance = initialI
    months = 0

#Here after going through the second interation of inputs from the lab 2 exercise, I noticed that the calculation was not giving the proper output because it was adding the interst payment even for the instant withdrawl of the investment, so it returned with 1 month and 30$
#As a solution I brought in a IF/Else statement that allows for both outputs to be excersized in the correct manner depending on the input of the user.
    if payout >= initialI:
        print("After 0 months your balance is {:.2f}".format(balance))
    else:
        while balance >= payout:
        #First for this calc we need to add the monthly interest to the balance - we can see that here by taking the balance times the monthly interest rate
            balance += balance * mrate
        #After we add the interst we have to subtract the payout from our intial investment (intitalI) which in this case is == to balance, for every itteration of the loop that we go through we have to add 1 month to show time has passed
            balance -= payout
            months +=1 
    #Finally we can structure our print output function to include the months and balance that were calculated above.
        print("After {} months your balance is {:.2f}.".format(months, balance))

#Lastly we call back the main function
main()
```

```python
#Advanced #3

#First thing we always do is define the main
def main():
    #Next we need to get all four of the user inputs in order to have the data required to make depreciation schedules. 
   #First we have a regular string input because of the text
    item = input("Please enter a description of the item: ")
    #Next we use float here because the cost of the item most likley will contain decimal places
    Ocost = float(input("\nPlease enter the cost of the item: "))
    #Here we use int because we dont want partial years for this example of depreciation, even though there are some deprecation schedules with half years included,
    life = int(input("\nPlease enter the estimated life of the item in whole years: "))
    #Lastly for inputs we use int again because we are defining before in the main funcito that the numbers 1 or 2 correspond with a specific method. 
    Mkind = int(input("\nPlease enter depreciation method (1 = straight line, 2 = double declining balance):"))
    
    #Next is the header and following data, this part took awhile as instead of using chatgpt to write the program I used the formatting with alignment matieral proided.  More explained in the dicussion.
    print("\nDepreciation schedule for: {}".format(item))
    #Here we can use the > and < operators to change the allignment and the number following can be seen as the width. We use < in the first alignment argument to get awawy from the left bar, from there we can use > to consistantly move a width of 10 spaced to the right from each last alignment argument.
    print("{:<5} {:>10} {:>10} {:>10}".format("Year","Begin","Dep","End"))

    #Next we have to print an intial row for the table, and we can do that by defining each of the columns we have assigned above, its almost like accumulator variables, where start is equal to our initial user input.
    start = Ocost
    print("{:<5} {:>10.2f} {:>10.2f} {:>10.2f}".format(0,start,0.00,start))

    #Next we have to write out the if statement that does the deprecation math depnding on the method chosen from the user input above. We can define the range with the user input "life" from above. We use the +1 in the second part of the function so we can capture the entire range and not end on the inputted number and we start at 1 because deprecation in year one is already accounted for in out earlier array formatting.
    #This function loops from year 1 through the life of the asset
    for year in range (1,life+1):
        #Here is the start of our nested if statement, this is letting the method chosen above by the user input either be deprecatied with straight line method or the double declining balance method.
        if Mkind == 1:
            #Here we have the fomula for straight line deprecation, one deduction a year
            dep = Ocost/ life
            #Here we have the double declining balance method, which is equivalent to double straight line because of 2/n. 
        elif Mkind == 2:
            rate = 2/life
              #Deprecation from current year times deprecation rate to find deprecation for that year
            dep = start * rate
            if year == life:
                #Here we are making sure the end of the depreciation schedule stops at 0 by just taking whats remaining as the last year of deprecitation instead of continuing to go negative. This only really applied to the double method because single line method is already evenly spaced apart.
                    dep = start
        #If no method variable above is met in the nested if statement, we can see the output "Invalid depreciation method below"           
        else:
            print("Invalid depreciation method.")
            return
        #Here we are subtracting the current years deprecation from the start (beginning balance) to find 'end' or what our asset is worth after depreceation at the end of the year
        end = start - dep
       
        #Here we repeat the array building function from above, but instead of using strings we have to use our newly defined variables otherwise our formatting is imcompatible.
        print("\n{:<5} {:>10} {:>10} {:>10}".format(year, start, dep, end))
        start = end
#Lastly we call the main function to execute the script above.
main()
```
